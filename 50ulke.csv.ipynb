```python
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.figsize':[9,9]},font_scale=1.2)
```

To Read The File by Pandas .

```python
df=pd.read_csv('50ulke.csv')
df.head()
```

To remove unvaluable data from dataset .

```python
df.drop(['Unnamed: 0'] , axis=1 , inplace=True)
```

The data is for 50 Countries and They located inside 6 Continents .

```python
df.info()
```

```python
df.describe()
```

To Turn The growth Value Into Positive

```python
df['Growth']=df['Growth'].apply(lambda x : abs(x))
df.describe()
```

To check if there any missing Data :

```python
df.isna().sum()
```

The most national Income values between 231 to 4000 and maximam National Income is 20412 and the lowest value is 231 .

The most countries growth rate between 1 and 4 .
maximam growth rate is 7.81 and the lowest growth rate is 0.53 .

```python
sns.kdeplot(df['Growth'],shade=True)
```
